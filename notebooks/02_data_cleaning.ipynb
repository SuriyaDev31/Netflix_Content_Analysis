```python
# 🧹 Step 5 – Part 2: Data Cleaning & Preprocessing

# 📘 1️⃣ Import Libraries and Load Raw Data
import pandas as pd
import numpy as np

# Load raw dataset
df = pd.read_csv(r"C:\Users\Dell\OneDrive\Documents\Projects\Netflix_Content_Analysis\data\raw\netflix_titles.csv")

# Preview data
print(df.head())

# 🔍 2️⃣ Handle Missing Values in other columns
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')

# 🧹 2️⃣a Robust Country Cleaning
# Convert all to string and strip spaces
df['country'] = df['country'].astype(str).str.strip()

# Replace invalid entries with NaN
invalid_countries = ['Unknown', '', 'N/A', '???', 'nan', 'NaN']
df['country'] = df['country'].replace(invalid_countries, pd.NA)

# Fix common typos
country_corrections = {
    'Untied States': 'United States',
    'Inda': 'India',
    'Uk': 'United Kingdom',
    'Usa': 'United States'
}
df['country'] = df['country'].replace(country_corrections)

# If multiple countries in one cell, keep only the first one
df['country'] = df['country'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else pd.NA)

# Drop rows where country is still missing
df = df.dropna(subset=['country'])

# Preview unique countries
print("Unique countries after robust cleaning:")
print(df['country'].unique())

# 🧹 3️⃣ Remove Duplicates
df.drop_duplicates(inplace=True)
print(f"Number of duplicates after cleaning: {df.duplicated().sum()}")

# 📅 4️⃣ Convert 'date_added' to Datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Create new columns: year_added and month_added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Check first rows
print(df[['date_added', 'year_added', 'month_added']].head())

# 🏷️ 5️⃣ Normalize Text Columns
df['title'] = df['title'].str.strip()
df['listed_in'] = df['listed_in'].str.strip()
df['country'] = df['country'].str.strip()

# 🧮 6️⃣ Optional: Feature Engineering
df['content_age'] = 2025 - df['release_year']  # Update current year accordingly
df['num_genres'] = df['listed_in'].apply(lambda x: len(x.split(', ')))

# 💾 7️⃣ Save Cleaned Data
df.to_csv(r"C:\Users\Dell\OneDrive\Documents\Projects\Netflix_Content_Analysis\data\processed\cleaned_data.csv", index=False)

print("Data cleaning complete. Cleaned file saved to cleaned_data.csv")
```
